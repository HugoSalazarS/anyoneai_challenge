```python
df = pd.read_csv("./sample_data/orders.csv")
df.head()
```

```python
# Question #1: How many unique orders are in the orders.csv file?

print('There are ', df['OrderID'].nunique(), 'unique orders in the file')
```

```python
# Question #2: What is the average number of items per order (rounded to two decimal places)?
average = df.groupby('OrderID').count()
print('the average number of ithems per order is: ',round(average['OrderTotal'].mean(),2))
```

```python
# Question #3: What is the highest number of items per order?
print('The highest number of items pero orden is: ', average['OrderTotal'].max())
```

```python
# Question #4: What is the number of orders placed in October 2021?
df['Date'] = pd.to_datetime(df['Date'])
october21 = df.query('Date >= "2021-10-01" & Date <= "2021-10-31"')
print('The numer of orders placed in October 21 is: ',october21['OrderID'].count())
```

```python
# Question #5: Which customer spent the most amount of money in 2021?
year21 = df.query('Date.dt.year == 2021')
year21.groupby('CustomerID')['Price'].sum().reset_index(name='total_spent').sort_values(by='total_spent', ascending=False)
```

The Customer who spent the most amount of money in 2021 was 5172443

```python
# Question #6: Historically, what is the best month for sales?
df.groupby(df['Date'].dt.strftime('%m'))['OrderTotal'].sum().reset_index(name='Total').sort_values(by='Total')
```

Historically, the best month for sales is December
